```python
import os
import torchaudio
import matplotlib.pyplot as plt

# 1️⃣ 指定 WAV 文件路径
file_path = "wav_from_parquet/19-198-0003.wav"

# 2️⃣ 获取文件大小（MB）
file_size_mb = os.path.getsize(file_path) / 1024 / 1024

# 3️⃣ 加载 WAV 文件
waveform, sample_rate = torchaudio.load(file_path)

# waveform: Tensor of shape [channels, samples]
num_channels, num_samples = waveform.shape
duration_sec = num_samples / sample_rate

# 4️⃣ 输出基础信息
print("=== WAV 文件信息 ===")
print(f"文件路径：{file_path}")
print(f"文件大小：{file_size_mb:.2f} MB")
print(f"采样率：{sample_rate} Hz")
print(f"通道数：{num_channels}")
print(f"总采样点数：{num_samples}")
print(f"音频时长：{duration_sec:.2f} 秒")
print(f"数据类型：{waveform.dtype}")  # <--- 查看数据类型（通常是 torch.float32）

# 5️⃣ 可选：绘制波形
plt.figure(figsize=(10, 4))
plt.title("Waveform")
plt.plot(waveform[0].numpy())  # 绘制第一个通道
plt.xlabel("Sample Index")
plt.ylabel("Amplitude")
plt.grid(True)
plt.show()
```
